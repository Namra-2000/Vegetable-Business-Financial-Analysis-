````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import warnings
warnings.filterwarnings('ignore')

# Load data
daily_df  = pd.read_csv('../data/daily_summary.csv')
daily_df['Date'] = pd.to_datetime(daily_df['Date'], dayfirst=True)
open_days = daily_df[daily_df['Business_Open'] == 'Yes'].copy()
print('Data loaded. Open business days:', len(open_days))
````

## 1. Revenue Formula Recap

```
Revenue = (40/100) × Vegetable_Cost + (60/100) × Sell_with_Margin
Profit  = Revenue − Vegetable_Cost − Other_Cost
```

> The owner sells **60% of stock at margin price** (5am–12pm) and clears **40% at cost price** to avoid wastage.

## 2. Vegetable Cost Analysis

````python
print('=== VEGETABLE COST STATISTICS ===')
print(open_days['Vegetables_Cost'].describe().round(2))
print(f'\nMode: ₹{open_days["Vegetables_Cost"].mode()[0]:,}')
````

````python
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(daily_df['Date'], daily_df['Vegetables_Cost'], color='#1d4ed8', linewidth=1.8)
ax.fill_between(daily_df['Date'], daily_df['Vegetables_Cost'], alpha=0.1, color='#1d4ed8')
ax.set_title('Total Vegetable Cost Over Time', fontsize=13, fontweight='bold')
ax.set_xlabel('Date'); ax.set_ylabel('Cost (₹)')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%b'))
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('../dashboards/veg_cost_trend.png', dpi=150)
plt.show()
````

## 3. Revenue Analysis

````python
print('=== REVENUE STATISTICS ===')
print(open_days['Revenue'].describe().round(2))
print(f'\nMode: ₹{open_days["Revenue"].mode()[0]:,}')
````

## 4. Daily Profit Trend

````python
print('=== PROFIT STATISTICS (Open Days Only) ===')
print(open_days['Profit'].describe().round(2))
print(f'\nMode: ₹{open_days["Profit"].mode()[0]:,}')
print(f'Sum (Total Monthly Profit): ₹{open_days["Profit"].sum():,.2f}')
````

````python
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(daily_df['Date'], daily_df['Profit'], color='#2563eb', linewidth=1.8, marker='o', markersize=4)
ax.axhline(y=open_days['Profit'].mean(), color='#f97316', linestyle='--', linewidth=1.2, label=f'Mean ₹{open_days["Profit"].mean():,.0f}')
ax.fill_between(daily_df['Date'], daily_df['Profit'], alpha=0.08, color='#2563eb')

# Annotate peak
peak = daily_df.loc[daily_df['Profit'].idxmax()]
ax.annotate(f'Peak\n₹{peak["Profit"]:,}\n(Republic Day eve)',
            xy=(peak['Date'], peak['Profit']),
            xytext=(peak['Date'], peak['Profit'] + 150),
            fontsize=8, color='#dc2626',
            arrowprops=dict(arrowstyle='->', color='#dc2626'))

ax.set_title('Daily Profit Over Time — Raj Kishore Gupta & Sons\nMubarakpur, Azamgarh', fontsize=13, fontweight='bold')
ax.set_xlabel('Date'); ax.set_ylabel('Profit (₹)')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%b'))
ax.legend()
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('../dashboards/profit_trend.png', dpi=150)
plt.show()
````

## 5. January vs February Profit Comparison

````python
jan_profit = open_days[open_days['Date'].dt.month == 1]['Profit'].mean()
feb_profit = open_days[open_days['Date'].dt.month == 2]['Profit'].mean()

print(f'January Average Profit : ₹{jan_profit:,.2f}')
print(f'February Average Profit: ₹{feb_profit:,.2f}')
print(f'Difference             : ₹{jan_profit - feb_profit:,.2f} (Jan was higher)')

fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(['January', 'February'], [jan_profit, feb_profit],
       color=['#1d4ed8', '#60a5fa'], edgecolor='white', width=0.4)
ax.set_title('Average Profit: January vs February', fontsize=12, fontweight='bold')
ax.set_ylabel('Avg Profit (₹)')
for i, v in enumerate([jan_profit, feb_profit]):
    ax.text(i, v + 10, f'₹{v:,.1f}', ha='center', fontweight='bold')
plt.tight_layout()
plt.savefig('../dashboards/jan_vs_feb.png', dpi=150)
plt.show()
````

## 6. Other Costs Trend

````python
print('Other Costs by Month:')
print(open_days.groupby(open_days['Date'].dt.month)['Other_Cost'].agg(['mean','min','max']))
print('\nNote: Jan=₹800 (transport+storage), Feb=₹850 (slight increase)')
````
