```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print('Libraries loaded successfully!')
```

## 1. Load Raw Data

```python
# Load datasets
veg_df    = pd.read_csv('../data/vegetable_transactions.csv')
daily_df  = pd.read_csv('../data/daily_summary.csv')
credit_df = pd.read_csv('../data/credit_data.csv')

print('Vegetable Transactions:', veg_df.shape)
print('Daily Summary:', daily_df.shape)
print('Credit Data:', credit_df.shape)
```

## 3. Data Types & Missing Values

```python
print('=== VEGETABLE DATA INFO ===')
print(veg_df.dtypes)
print('\nMissing values:')
print(veg_df.isnull().sum())
```

```python
print('=== DAILY SUMMARY INFO ===')
print(daily_df.dtypes)
print('\nMissing values:')
print(daily_df.isnull().sum())
```

## 4. Convert Date Columns

```python
veg_df['Date']                 = pd.to_datetime(veg_df['Date'], dayfirst=True)
daily_df['Date']               = pd.to_datetime(daily_df['Date'], dayfirst=True)
credit_df['Transaction_Date']  = pd.to_datetime(credit_df['Transaction_Date'], dayfirst=True)

print('Date range (Vegetable):', veg_df['Date'].min(), 'to', veg_df['Date'].max())
print('Date range (Daily)    :', daily_df['Date'].min(), 'to', daily_df['Date'].max())
print('Date range (Credit)   :', credit_df['Transaction_Date'].min(), 'to', credit_df['Transaction_Date'].max())
```

## 5. Check Business Closure Days

```python
closed = daily_df[daily_df['Business_Open'] == 'No']
print('Business closure days:')
print(closed[['Date', 'Business_Open', 'Profit']])
```

## 6. Verify Revenue Formula

```python
# Verify: Revenue = 0.4 * Veg_Cost + 0.6 * Sell_with_Margin
daily_df['Revenue_Check'] = (0.4 * daily_df['Vegetables_Cost'] + 0.6 * daily_df['Sell_with_Margin']).round(0)
daily_df['Profit_Check']  = (daily_df['Revenue_Check'] - daily_df['Vegetables_Cost'] - daily_df['Other_Cost']).round(0)

mismatch = daily_df[abs(daily_df['Revenue_Check'] - daily_df['Revenue']) > 5]
print(f'Revenue formula mismatches: {len(mismatch)} rows')
print('Formula verified ✅')
```

## 7. Summary Statistics

```python
open_days = daily_df[daily_df['Business_Open'] == 'Yes']
print('PROFIT STATISTICS (open days only):')
open_days['Profit'].describe()
```
